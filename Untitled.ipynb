```python
import numpy as np
import pandas as pd
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.models import Sequential
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
import cv2
```

```python
path = "../posterPrediction/cleanCsv.csv"
df = pd.read_csv(path)
del df['Unnamed: 0']
df
```

```python
ids = df['imdbId']
imgs = []
for id in ids:
    img = cv2.imread("posters/%d.jpg" % id)
    (b,g,r) = cv2.split(img)
    img=cv2.merge([r,g,b])
    imgs.append(img)
    
%matplotlib inline
plt.imshow(imgs[1])
```

```python
def singleGenre(genres):
    strings = str(genres)
    single = strings.split("|")[0]
    return single
df['Genre'] = df['Genre'].apply(singleGenre)
```

```python
uniques = []
for x in df['Genre']:
    if x not in uniques:
        uniques.append(x)
        
def freqFinder(gen):
    count = 0
    for x in df['Genre']:
        if x == gen:
            count += 1
    print(gen, count)

for x in uniques:
    freqFinder(x)
```
