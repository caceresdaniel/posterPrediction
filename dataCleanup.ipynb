```python
import numpy as np
import pandas as pd
```

```python
path = "../posterPrediction/MovieGenrefull.csv"
df = pd.read_csv(path, encoding='latin-1')
df1 = pd.read_csv(path, encoding='latin-1')
del df["Imdb Link"]
del df['IMDB Score']
del df['Poster']
df = df.drop_duplicates()
```

```python
with open('../posterPrediction/files.txt') as f:
  lineList = f.readlines()
clean = []
for x in lineList:
    clean.append(int(x.strip(".jpg\n")))

ids = df['imdbId']

delet = []

for x in ids:
    if int(x) not in clean:
        delet.append(x)
```

```python
for x in delet:
    df = df.drop(df.loc[df['imdbId'] == x].index)
```
